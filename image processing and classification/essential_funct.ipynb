```python
import cv2 as cv
from matplotlib import pyplot
```

```python
img = cv.imread('Resources/Images/cat.jpg')
img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_blur = cv.GaussianBlur(img, (7, 7), cv.BORDER_DEFAULT) # al doilea param pentru a modifica 'bluritate'
img_canny = cv.Canny(img, 125, 175)
blur_img_canny = cv.Canny(img_blur, 125, 175)
dilate_img = cv.dilate(blur_img_canny, (7, 7), iterations=3)
```

```python
fig, axs = pyplot.subplots(2, 3, constrained_layout = True)
axs[0, 0].imshow(img)
axs[0, 0].set_title('Normal Image') 
axs[0, 0].axis('off')

axs[0, 1].imshow(img_grey)
axs[0, 1].set_title('Gray Scaled Image')
axs[0, 1].axis('off')

axs[1, 1].imshow(blur_img_canny)
axs[1, 1].set_title('Blured Canny Image') 
axs[1, 1].axis('off')

axs[0, 2].imshow(img_blur)
axs[0, 2].set_title('Blurred Image') 
axs[0, 2].axis('off')

axs[1, 0].imshow(img_canny)
axs[1, 0].set_title('Canny Image') 
axs[1, 0].axis('off')

axs[1, 2].imshow(dilate_img)
axs[1, 2].set_title('Dilated Image') 
axs[1, 2].axis('off')
```
