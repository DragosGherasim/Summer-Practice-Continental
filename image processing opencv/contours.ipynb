```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Resources\Images\cats.jpg')
grey_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
blur_img = cv.GaussianBlur(img, (7, 7), cv.BORDER_DEFAULT)
canny_img = cv.Canny(blur_img, 125, 175)
ret, thresh = cv.threshold(grey_img, 125, 255, cv.THRESH_BINARY)
```

```python
contours_thresh, hierarchies_thresh = cv.findContours(thresh, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
contours_canny, hierarchies_canny = cv.findContours(canny_img, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
```

```python
fig, axis = plt.subplots(1, 2, figsize = (15, 15))
axis[0].imshow(img)
axis[0].axis("off")
axis[0].set_title('Normal Image')

axis[1].imshow(thresh)
axis[1].axis("off")
axis[1].set_title('Thresh Image')
```
