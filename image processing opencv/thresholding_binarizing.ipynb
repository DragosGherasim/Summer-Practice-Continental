```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Resources/Images/cats.jpg')
img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
thresh, thresh_img = cv.threshold(img_grey, 127, 255, cv.THRESH_BINARY)
```

```python
fig, axis = plt.subplots(1, 2, figsize = (15, 15))
axis[0].imshow(img_grey, cmap = 'gray')
axis[0].axis('off')
axis[0].set_title('Gray-Scaled')

axis[1].imshow(thresh_img, cmap = 'gray')
axis[1].axis('off')
axis[1].set_title('Binarizaion')
```
